```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get(url="https://en.wikipedia.org/wiki/Cristiano_Ronaldo")
page.status_code
```

```python
soup = BeautifulSoup(page.content,"html.parser")
title = soup.find("h1",id="firstHeading")
print(title)
```

```python
title.text
```

```python
data = soup.find("div",id="mw-content-text")
print(data.text)
```

```python
data = pd.read_excel("WebScrapFootball.xlsx")
data
```

```python
data['Names']
```

```python
type(data['Names'])
```

```python
data['Names']= data['Names'].str.replace(' ','_')
```

```python
data["Names New"]
```

```python
def change(val):
    val2=val.split(" ")
    val2="_".join(val2)
    print(val2)
    return val2
```

```python
data["Names New"] = data['Names'].apply(change)
```

```python
data.head
```

```python
data_list=list(data["Names New"])
```

```python
data_list[1]
```

```python
titles=[]
Info=[]
for i in data_list:
    page = requests.get(f'https://en.wikipedia.org/wiki/{i}')
    soup = BeautifulSoup(page.content,"html.parser")
    title = soup.find("h1",id="firstHeading").text
    content = soup.find("div", id="mw-content-text").text
    titles.append(title)
    Info.append(content)
    
print(titles)
```

```python
all_Data = {"Names":titles,"Data":Info}
all_Data
```

```python
df = pd.DataFrame(all_Data)
df.to_csv('Scrapped_Data.csv')
```
